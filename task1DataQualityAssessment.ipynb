## Merge all information tables

```python
demog['customer_id'].iloc[-1]
```

```python
# add id to new customer
newcus.insert(0, 'customer_id', range(4001, 4001 + len(newcus)))
```

```python
newcus
```

```python
#  merge the Customer Demographic with the Customer Address using the customer_id column

customer_demographic = pd.merge(demog, addr, how='left', on='customer_id')

```

```python

customer_demographic
```

```python
new_df = pd.concat([customer_demographic, newcus], ignore_index=True, sort=False)
```

```python
customer_demographic = new_df
```

```python
customer_demographic.postcode = customer_demographic.postcode.astype(str).apply(lambda x: x.replace('.0',''))
```

```python
# clean the data
customer_demographic['gender'] = customer_demographic['gender'].replace('F','Female').replace('M','Male').replace('Femal','Female').replace('U','Unspecified')
```
